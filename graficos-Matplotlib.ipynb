### Grafico de Histograma 

```python
import pandas as pd
wine = pd.read_csv("D:\Data_Science_With_Python\databases\wine_dataset.csv")
display(wine)
```

```python
import matplotlib.pyplot as plt
```

```python
wine.hist(column = 'pH', bins=10)
plt.show()
```

```python
wine.hist(column = "alcohol", bins=10)
plt.show()
```

### Boxplot 

```python
import matplotlib.pyplot as plt

wine.boxplot(column = 'alcohol') #cria um grafico
plt.show()
```

```python
wine.boxplot(column = ['fixed_acidity','volatile_acidity', 'citric_acid']) #cria mais de um grafico
```

### Vários gráficos ao mesmo tempo 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1000, 1000, 1)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
y1 = x**2
y2 = x**3
y3 = -x**2
y4 = -x**3
plt.subplot(2,2,1)
plt.plot(x, y1)
```
